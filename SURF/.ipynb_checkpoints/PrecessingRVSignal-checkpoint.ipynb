```python
from radvel import kepler
import numpy as np

from matplotlib import pyplot as plt

period = 10
tp =1
e = 0.4
om = 2
K = 10
times = np.linspace(0,40,200)
Tprecess = 50 * period
omegadot = 2 * np.pi /Tprecess

rv_signal = kepler.rv_drive(times,[period,tp,e,om,K])

rv_precession_signal = [
    kepler.rv_drive(np.array([t]),
                    [period,tp,e,om + t*omegadot,K]) 
    for t in times]
plt.plot(times,rv_signal)
plt.plot(times,rv_precession_signal)
```
